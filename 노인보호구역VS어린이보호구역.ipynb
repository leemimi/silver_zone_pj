```python
import warnings
warnings.filterwarnings(action='ignore') 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib
import platform
import seaborn as sns 
import matplotlib
matplotlib.rcParams['font.family']
if platform.system() == 'Windows':
    matplotlib.rc('font', family='Malgun Gothic')
    #Malgun 
%matplotlib inline
```

```python
df1 = pd.read_csv('동대문구_어린이보호구역_개수.csv',encoding='euc-kr')
df2 = pd.read_csv('동대문구_노인보호구역위치.csv',encoding='euc-kr')
```

```python
df1
```

```python
count1 = sum(df1.index)
count2=sum(df2['장소유형코드'])
```

```python
count = [count1, count2]
```

```python
all= pd.DataFrame(count)
```

```python
all.columns = ['갯수']
```

```python
all.rename(index={0:'동대문구_어린이보호구역',1:'동대문구_노인보호구역'},inplace=True)
```

```python
all
```

```python
data= all['갯수']
colors = ['#FFA2AD','#6478FF']
wedgeprops = {'width':0.7, 'edgecolor':'w','linewidth':5 }
plt.title('동대문구_노인보호구역과 노인 복지시설 갯수')
plt.pie(data, labels=all.index, autopct='%.1f%%',startangle=260, counterclock=False,colors=colors, wedgeprops=wedgeprops)
plt.show()
```
